## 1.1: Capitalización de Mercado actual - Top Cryptomonedas:

### A continuación se vestudiarán los datos y gráficos sobre el valor y posición de Bitcoin y Ethereum respecto al resto de las cryptomonedas, para estudiar su relación con el mercado, el dominio de estas y la magnitud de la capitalización de mercado total de las criptomonedas.

```python
Market_Cap_Total = top200crypto["Market Cap"].sum()
Market_Cap_sansBTC = top200crypto["Market Cap"].iloc[1:-1].sum()
Market_Cap_BTC = top200crypto["Market Cap"][0]
Market_Cap_ETH = top200crypto["Market Cap"][1]
porc_MC_BTC = Market_Cap_BTC*100/Market_Cap_Total
porc_MC_ETH = Market_Cap_ETH*100/Market_Cap_Total
porc_MC_RESTO = 100-porc_MC_BTC-porc_MC_ETH
```

```python
print("Cuota de Mercado Total:","$"+str(Market_Cap_Total),"----", "Cuota de Mercado Total sin Bitcoin:","$"+str(Market_Cap_sansBTC))
print("Cuota de Mercado Bitcoin:","$"+str(Market_Cap_BTC),"----", "Cuota de Mercado Ethereum:","$"+str(Market_Cap_ETH))
print("Porcentaje de Cuota de Mercado de Bitcoin:", str(porc_MC_BTC) + "%")
print("Porcentaje de Cuota de Mercado de Ethereum:", str(porc_MC_ETH) + "%")
print("Porcentaje de Cuota de Mercado de Ethreum sobre Bitcoin:", str(Market_Cap_ETH*100/Market_Cap_BTC) + "%")
```

```python
labels = ['Bitcoin','Ethereum','Resto del Mercado']
values = [porc_MC_BTC, porc_MC_ETH, porc_MC_RESTO]

grafico1_1 = px.pie(values=values, names=labels, title='Capitalización de Mercado % BTC, ETH y otros',
color_discrete_sequence=px.colors.sequential.Aggrnyl,width=600)
grafico1_1.update_traces(textposition='inside', textinfo='percent+label')
grafico1_1.show()
grafico1_1.write_html("Utils\Visualizations\MarketCap_BTC_ETH_RESTO.html")
#Guardamos el gráfico en una librería de Chart Studio, para poder usarlo en presentaciones.
#py.plot(grafico1_1, filename="MarketCapBTCETHRESTO", auto_open=False)

```

```python
months = ["May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Jan","Feb","Mar","Apr"] #se declaran las variables de los ejes x,y
years = [str(x) for x in range(2017,2022)] #por las limitaciones en los datos de algunas cryptomonedas, el gráfico inicia en Mayo 2017
                                        #y termina en Septiembre 2021.

#para los valores "z" de cada variable tomaremos en cuenta los cambios en los valores porcentuales 
# y declararemos variables mensuales que tomen el último día de cada mes. Luego se recorren los precios(%) invertidos
#durante los 12 meses de cada año y se aplica la función trp del módulo de funciones.
xrp_mensual = crypto2017_2021["Ripple"].resample("1M").last() 
xrp_mensual = xrp_mensual.pct_change()*100
z_vals_xrp = [list(xrp_mensual["Prices"].iloc[::-1])[i: i+12] for i in range(0,len(xrp_mensual),12)]
z_vals_xrp = [fp.trp(x,12)for x in z_vals_xrp]

ltc_mensual = crypto2017_2021["Litecoin"].resample("1M").last()
ltc_mensual = ltc_mensual.pct_change()*100
z_vals_ltc = [list(ltc_mensual["Prices"].iloc[::-1])[i: i+12] for i in range(0,len(ltc_mensual),12)]
z_vals_ltc = [fp.trp(x,12)for x in z_vals_ltc]

eth_mensual = crypto2017_2021["Ethereum"].resample("1M").last()
eth_mensual = eth_mensual.pct_change()*100
z_vals_eth = [list(eth_mensual["Prices"].iloc[::-1])[i: i+12] for i in range(0,len(eth_mensual),12)]
z_vals_eth = [fp.trp(x,12)for x in z_vals_eth]

btc_mensual = crypto2017_2021["Bitcoin"].resample("1M").last()
btc_mensual = btc_mensual.pct_change()*100
z_vals_btc = [list(btc_mensual["Prices"].iloc[::-1])[i: i+12] for i in range(0,len(btc_mensual),12)]
z_vals_btc = [fp.trp(x,12)for x in z_vals_btc]

doge_mensual = crypto2017_2021["Dogecoin"].resample("1M").last()
doge_mensual = doge_mensual.pct_change()*100
z_vals_doge = [list(doge_mensual["Prices"].iloc[::-1])[i: i+12] for i in range(0,len(doge_mensual),12)]
z_vals_doge = [fp.trp(x,12)for x in z_vals_doge]

usdt_mensual = crypto2017_2021["Tether"].resample("1M").last()
usdt_mensual = usdt_mensual.pct_change()*100
z_vals_usdt = [list(usdt_mensual["Prices"].iloc[::-1])[i: i+12] for i in range(0,len(usdt_mensual),12)]
z_vals_usdt = [fp.trp(x,12)for x in z_vals_usdt]

grafico6 = make_subplots(rows=2, cols=3,subplot_titles=("Bitcoin", "Ethereum", "Ripple", "Dogecoin", "Litecoin", "Tether"))

grafico6.add_trace(go.Heatmap(z=z_vals_xrp,x=months,y=years,zmax=100,zmin=-50), row=1, col=3)

grafico6.add_trace(go.Heatmap(z=z_vals_ltc,x=months,y=years,zmax=100,zmin=-50), row=2, col=2)

grafico6.add_trace(go.Heatmap(z=z_vals_eth,x=months,y=years,zmax=100,zmin=-50), row=1, col=2)

grafico6.add_trace(go.Heatmap(z=z_vals_btc,x=months,y=years,zmax=100,zmin=-50), row=1, col=1)

grafico6.add_trace(go.Heatmap(z=z_vals_doge,x=months,y=years,zmax=100,zmin=-50), row=2, col=1)

grafico6.add_trace(go.Heatmap(z=z_vals_usdt,x=months,y=years,zmax=100,zmin=-50), row=2, col=3)

grafico6.update_layout(height=600, width=1000, title_text="Variación % Precio-Mes Cryptomonedas 2017-2021")
grafico6.show()
grafico6.write_html("Utils\Visualizations\Estacionalidadcrypto.html")
#Guardamos el gráfico en una librería de Chart Studio, para poder usarlo en presentaciones.
#py.plot(grafico6, filename="estacionalidadcrypto", auto_open=False)
```
